```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report, f1_score
import lightgbm as lgb
from collections import Counter
import warnings
import pickle
from catboost import CatBoostClassifier
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('is_train_data/is_train_20190701.txt')
```

```python
pro = pd.read_csv('per_road_pro.txt')
train = pd.merge(train, pro, how='left')
```

```python
y = train.label
train = train.drop('label',1)
train = train.drop('is_high',1)
```

```python
train_1 = pd.read_csv('is_train_data//is_train_20190730.txt')
y_5 = train_1.label
train_1 = pd.merge(train_1, pro, how='left')
train_1 = train_1.drop('label',1)

train_1 = train_1.drop('is_high',1)
```

```python
s = train.columns.values
```

```python
max_s = 0.5619985255427103
```

```python
s = train.columns.values
for i in tqdm(range(s.size)):
    
    ctrain = train.drop(labels = s[i], axis = 1)
    ctrain_1 = train_1.drop(labels = s[i], axis = 1)
    
    X_train, X_validation, y_train, y_validation = train_test_split(ctrain,y,test_size=0.2 , random_state=1234)
    classweight = {0:0.2,1:0.2,2:0.6}

    model = CatBoostClassifier(iterations=500, depth=5,early_stopping_rounds=20,learning_rate=0.5, loss_function='MultiClass',
                                logging_level='Verbose',class_weights=classweight)

    model.fit(X_train,y_train,eval_set=(X_validation,y_validation),verbose=False)
    
    report = f1_score(y_5, model.predict(ctrain_1), average=None)
    Score = report[0] * 0.2 + report[1] * 0.2 + report[2] * 0.6
    print('Score: ', report[0] * 0.2 + report[1] * 0.2 + report[2] * 0.6)
    if (max_s < Score):
        max_s = Score
        train = ctrain
        train_1 = ctrain_1
```
